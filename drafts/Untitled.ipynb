```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
π = np.pi
```

```python
X = np.linspace(-π, π, 100).reshape(-1, 1)
ϵ = np.random.laplace(size=X.shape)
α = 10
y = α*np.sin(X) + ϵ
p = LinearRegression().fit(X, y).predict(X)

plt.scatter(X, y);
plt.plot(X, p)
```

```python
plt.hist(y - p)
```
